```python
'''This tutorial will help you understand the basic image related operations like loading an image, resizing an image,
cropping and rotating an image etc'''


from PIL import Image #Python Imaging Library(PIL) - Pillow
import numpy as np
import matplotlib.pyplot as plt

# 1. Load an image
img = Image.open('D:\CourseIITK\EE200\EndTermProject\EE200_practical_programming\cat_gray.jpg')  # Replace with your file path

# 2. Display original image
plt.figure(figsize=(6, 4))
plt.imshow(img)
plt.title("Original Image")
plt.axis('off')
plt.show()

# 3. Convert to grayscale
img_gray = img.convert('L') #L represents luminance meaning a number b/w 0-225 represnting brightness

# 4. Resize image
img_resized = img.resize((50, 50)) # scales an image to 50 pixels wide and 50 pixels high # by default used Image.BICUBIC as argument for RGB images

# 5. Crop image (left, upper, right, lower)
img_cropped = img.crop((50, 50, 200, 200)) #left - x of left edge; right - x of right edge; upper - y of upper edge; lower - y of lower edge

# 6. Rotate image
img_rotated = img.rotate(45)  # Rotate 45 degrees counter-clockwise

# 7. Display processed images in a single plot using subplot function
fig, axs = plt.subplots(1, 4, figsize=(16, 4))

axs[0].imshow(img_gray, cmap='gray')
axs[0].set_title("Grayscale")
axs[0].axis('off')

axs[1].imshow(img_resized)
axs[1].set_title("Resized")
axs[1].axis('off')

axs[2].imshow(img_cropped)
axs[2].set_title("Cropped")
axs[2].axis('off')

axs[3].imshow(img_rotated)
axs[3].set_title("Rotated")
axs[3].axis('off')

plt.tight_layout()
plt.show()
```

```python
'''This tutorial will help you understand the basic audio related operations like loading an audio, fetching sampling rate and duration,
plotting the waveform and the spectrogram etc'''
import librosa
import librosa.display
import numpy as np
import matplotlib.pyplot as plt

# 1. Load the audio file
audio_path = "D:\CourseIITK\EE200\EndTermProject\EE200_practical_programming\song_with_2piccolo.wav"
y, sr = librosa.load(audio_path, sr=None) # sr stands for sampling rate

# 2. Print sampling rate and duration
print(f"Sample Rate: {sr} Hz")
print(f"Duration: {len(y) / sr:.2f} seconds")

# 3. Normalize audio
y_normalized = y / np.max(np.abs(y))

# 4. Plot the waveform
plt.figure(figsize=(10, 4))
librosa.display.waveshow(y_normalized, sr=sr)
plt.title("Waveform")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.tight_layout()
plt.show()

# 5. Plot the spectrogram (in dB)
D = librosa.stft(y)
S_db = librosa.amplitude_to_db(np.abs(D), ref=np.max)

plt.figure(figsize=(10, 4))
librosa.display.specshow(S_db, sr=sr, x_axis='time', y_axis='hz', cmap='magma')
plt.colorbar(format="%+2.0f dB")
plt.title("Spectrogram")
plt.tight_layout()
plt.show()
```
